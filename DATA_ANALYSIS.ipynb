# DATA_ANALYSIS

```python
import pandas as pd
```

```python
# 1. In how many cities Summer Olympics is held so far?
```

```python
df=pd.read_csv('Summer_data.csv')
print(len(df['City'].unique()))
```

```python
# 2. Which sport is having most number of Gold Medals so far? (Top 5)

c=df[df['Medal']=='Gold']
p=[]
for i in c['Sport'].unique():
    p.append([i,len(c[c['Sport']==i])])
p=pd.DataFrame(p,columns=['sport','freq'])
p=p.sort_values(by="freq",ascending=False)
p=p.head()
print(p)
p.plot(x='sport',y='freq',kind='bar')
```

```python
#3. Which sport is having most number of medals so far? (Top 5)

x=[]
y=df[['Sport','Medal']]
y=y.dropna()
for i in y['Sport'].unique():
    x.append([i,len(y[y['Sport']==i])])
x=pd.DataFrame(x,columns=['sport','freq'])
x=x.sort_values(by="freq",ascending=False)
x=x.head()
print(x)
x.plot(x='sport',y='freq',kind='bar')
```

```python
# 4. Which player has won most number of medals? (Top 5)

z=df[['Athlete','Medal']]
z=z.dropna()
q=[]
for i in z['Athlete'].unique():
    q.append([i,len(z[z['Athlete']==i])])
q=pd.DataFrame(q,columns=['Athlete','freq'])
q=q.sort_values(by="freq",ascending=False)
q=q.head()
print(q)
q.plot(x='Athlete',y='freq',kind='bar')
```

```python
# 6. In which year India won first Gold Medal in Summer Olympics?

u=df[['Year','Country','Medal']]
u=u.dropna()
u=u[u['Medal']=='Gold']
u=u[u['Country']=='IND']
u=u.sort_values(by="Year")
u=u.head(1)
print("year=",u['Year'].values)
```

```python
# 7. Which event is most popular in terms on number of players? (Top 5)

v=df[['Athlete','Event']]
v=v.dropna()
o=[]
m=[]
for i in v['Event'].unique():
    n=v[v['Event']==i]
    for j in n['Athlete'].unique():
        o.append([i,j])
o=pd.DataFrame(o,columns=['Event','Athlete'])
for i in o['Event'].unique():
    m.append([i,len(o[o['Event']==i])])
m=pd.DataFrame(m,columns=['Event','freq'])
m=m.sort_values(by="freq",ascending=False)
m=m.head()
print(m)
m.plot(x='Event',y='freq',kind='bar')
```

```python
# 8. Which sport is having most female Gold Medalists? (Top 5)

t=df[['Sport','Gender','Medal']]
t=t.dropna()
t=t[t['Medal']=='Gold']
t=t[t['Gender']=='Women']
s=[]
for i in t['Sport'].unique():
    s.append([i,len(t[t['Sport']==i])])
s=pd.DataFrame(s,columns=['sport','freq'])
s=s.sort_values(by="freq",ascending=False)
s=s.head()
print(s)
s.plot(x='sport',y='freq',kind='bar')
```
